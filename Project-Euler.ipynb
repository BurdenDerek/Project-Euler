```python
# final draft

def special_pythagorean_triplet():
    # this function is looking for the first pythagorean triplet
    # where a + b + c = 1000
    # then returns a*b*c for that triplet
    
    # while avoiding single letter variables is good practice in general
    # i feel as a, b, c are the most appropriate because the formula is a^2 + b^2 = c^2
    
    # this nested loop is to build all possible a b combinations
    # the a range is arbitrary high number to get us to the answer
    # the b range is based on the current a to prevent double checking combinations
    for a in range(1000):
        for b in range(a):
            # solve for c, a^2 + b^2 = c^2, so c equals the square root of (a^2 + b^2)
            c = (a**2 + b**2)**.5
            # c needs to be an interger for it to be a triplet
            # and the triples sum needs to equal 1000
            if (c == int(c)) & (a + b + c == 1000):
                # return the product of a, b, c
                # int the product because the float is an artifact from the square root
                return int(a*b*c)
        
print(special_pythagorean_triplet())
```
